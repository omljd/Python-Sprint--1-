#### Plot Number of Seats Won by Each Party

#### Get the votes for Rahul Gandhi, Narendra Modi, and Amit Shah 

```python
rahul_entries = data[data['Leading Candidate'] == 'RAHUL GANDHI']
modi_entries = data[data['Leading Candidate'] == 'NARENDRA MODI']
amit_entries = data[data['Leading Candidate'] == 'AMIT SHAH']

# Get the votes for Rahul Gandhi, Narendra Modi, and Amit Shah
rahul_votes = rahul_entries['Margin'].values
modi_votes = modi_entries['Margin'].values[0] if not modi_entries.empty else 0
amit_votes = amit_entries['Margin'].values[0] if not amit_entries.empty else 0

# Get the original constituency names for Rahul Gandhi
rahul_constituencies = list(rahul_entries['Constituency'])

# Get the original constituency name for Narendra Modi
modi_constituency = modi_entries['Constituency'].values[0] if not modi_entries.empty else "Modi Constituency"

# Get the original constituency name for Amit Shah
amit_constituency = amit_entries['Constituency'].values[0] if not amit_entries.empty else "Amit Shah Constituency"

# Combine the data
data_to_plot = pd.DataFrame({
    'Candidate': ['Rahul Gandhi'] * len(rahul_votes) + ['Narendra Modi', 'Amit Shah'],
    'Constituency': rahul_constituencies + [modi_constituency, amit_constituency],
    'Votes': list(rahul_votes) + [modi_votes, amit_votes]
}) 

# Plot the comparison
plt.figure(figsize=(12, 6))
sb.barplot(data=data_to_plot, x='Constituency', y='Votes', hue='Candidate', palette='muted')
plt.title('Comparison of Votes for Rahul Gandhi, Narendra Modi, and Amit Shah')
plt.xlabel('Constituency')
plt.ylabel('Votes')
plt.xticks(rotation=45)
plt.show()
```


### Insight:
The bar plot compares the votes received by Rahul Gandhi, Narendra Modi, and Amit Shah across their respective constituencies, highlighting the electoral performance of these prominent candidates and revealing which constituencies favored each candidate.

#### Highest and lowest Victory Candidate

```python
highest_margin_entry = data.loc[data['Margin'].idxmax()]
lowest_margin_entry = data.loc[data['Margin'].idxmin()]

#Combine the data
data_to_plot = pd.DataFrame({
    'Candidate': [highest_margin_entry['Leading Candidate'], lowest_margin_entry['Leading Candidate']],
    'Party': [highest_margin_entry['Leading Party'], lowest_margin_entry['Leading Party']],
    'Margin': [highest_margin_entry['Margin'], lowest_margin_entry['Margin']]
})

# Plot the comparison
plt.figure(figsize=(10, 6))
sb.barplot(data=data_to_plot, x='Candidate', y='Margin', hue='Party', palette='muted')
plt.title('Candidates with Highest and Lowest Margin of Victory')
plt.xlabel('Candidate')
plt.ylabel('Margin of Victory (')
plt.xticks(rotation=45)
plt.show()
```


### Insight:
The bar plot illustrates the candidates with the highest and lowest margins of victory, along with their respective parties, showcasing the electoral extremes in terms of vote margins and highlighting significant performance disparities among candidates.

#### Win Distribution by Party

```python
party_votes = data.groupby('Leading Party')['Margin'].sum().sort_values(ascending=False)

# Plot pie chart
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(party_votes, labels=None, autopct='%1.1f%%', startangle=140, wedgeprops=dict(edgecolor='w'))
plt.title('Win Distribution by Party', pad=20)
plt.axis('equal')

plt.legend(labels=party_votes.index, loc='center left', bbox_to_anchor=(1, 0.5), fontsize='medium')


plt.show()
```

### Insight: 
The pie chart displays the distribution of total winning margins by party, illustrating which parties secured the largest portion of overall victories.

#### Analysis Top 10 Trailing party

```python
#Calculates the total trailing margin for each party and sorts them in descending order.

trailing_party_votes=data.groupby('Trailing Party')['Margin'].sum().sort_values(ascending=False)
#Counts the number of times each party appears in the 'Trailing Party' column, which effectively counts the number of seats where each party is trailing.
trailing_party_seats=data['Trailing Party'].value_counts()
```

### Insight: 
The analysis shows the total trailing margins for each party and the number of seats where they were trailing, highlighting their electoral challenges.

```python
#Initializes a figure for plotting, with a size of 20 inches wide and 6 inches tall. 
plt.figure(figsize=(20, 6))

# Plot votes distribution by trailing party
# Creates a subplot in a grid of 1 row and 2 columns, selecting the first subplot position (left side). This is useful when planning to create multiple plots side-by-side.
plt.subplot(1, 2, 1)
sb.barplot(
    x=trailing_party_votes.index[:10], 
    y=trailing_party_votes.values[:10], 
    palette='viridis', 
    hue=trailing_party_votes.index[:10],  # Assigning the x variable to hue
    legend=False  # Set legend to False
)
plt.title('Top 10 Trailing Parties ')
plt.xlabel('Party')
plt.ylabel('Total Votes')
#Rotates the x-axis labels by 45 degrees to improve readability, especially useful when party names are long or when there are many bars close together.
plt.xticks(rotation=45)
```

### Insight: 
The bar plot illustrates the top 10 trailing parties based on total votes, providing a clear view of which parties are facing electoral challenges and how they compare in terms of trailing vote totals.
